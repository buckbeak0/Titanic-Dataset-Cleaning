```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Titanic-Original.csv')
print(df.describe())
print(df.info())
```

```python
df.isnull().sum()
```

```python
df.drop(['Cabin', 'Ticket', 'Name'], inplace=True, axis=1)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df.head()
```

```python
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
df['Embarked'] = df['Embarked'].map({'S':0, 'C':1, 'Q':2})
df.head()
```

```python
df['FamliySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = 1
df['IsAlone'].loc[df['FamliySize'] > 1] = 0
```

```python
df.info()
df.head()
```

```python
df.to_csv('Titanic-Cleaned.csv', index=False)
```
